```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
model_data=pd.read_csv(r'C:\Users\amikula pavani\jupyter-pycodes\mlproject\data\salesML.csv')
```

```python
model_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = model_data.drop(["Sales", "Customers"], axis=1)
y = model_data["Sales"]
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
ypred=lr.predict(xtest)
```

## Evaluation Metrics

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mae=mean_absolute_error(ytest,ypred)
mse=mean_squared_error(ytest,ypred)
rmse=mse**0.5
r2=r2_score(ytest,ypred)

print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
```

```python
dt.fit(xtrain,ytrain)
```

```python
ypred_dt=dt.predict(xtest)
```

```python
mae_dt = mean_absolute_error(ytest, ypred_dt)
mse_dt = mean_squared_error(ytest, ypred_dt)
rmse_dt = mse_dt ** 0.5
r2_dt = r2_score(ytest, ypred_dt)

print("MAE :", mae_dt)
print("MSE:", mse_dt)
print("RMSE:", rmse_dt)
print("R2 Score:", r2_dt)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

rf.fit(xtrain, ytrain)
```

```python
ypred_rf=rf.predict(xtest)
```

```python
mae_rf = mean_absolute_error(ytest, ypred_rf)
mse_rf = mean_squared_error(ytest, ypred_rf)
rmse_rf = mse_rf ** 0.5
r2_rf = r2_score(ytest, ypred_rf)

print("MAE :", mae_rf)
print("MSE:", mse_rf)
print("RMSE:", rmse_rf)
print("R2 Score:", r2_rf)
```

## Conclusion:

* Multiple machine learning models were trained and evaluated for retail sales forecasting.
* Linear Regression served as a baseline model but showed limited predictive capability (R² ≈ 0.22). 
* Decision Tree significantly improved performance (R² ≈ 0.82).
* XGBoost achieved further improvement (R² ≈ 0.85). 
* Random Forest delivered the best performance with an R² score of approximately 0.91, along with the lowest MAE and RMSE. 
*Therefore, Random Forest was selected as the final model for sales prediction.
