## Feature Engineering

```python
modeldata['Date'].dtype
```

* We can't apply date operations on string data type so, we convert it into Datetime dtype.

```python
modeldata['Date']=pd.to_datetime(modeldata['Date'])
```

```python
modeldata['Date'].dtype
```

### Extracting useful features

```python
modeldata['Year']=modeldata['Date'].dt.year
modeldata['Month']=modeldata['Date'].dt.month
modeldata['Day']=modeldata['Date'].dt.day
modeldata['WeekOfYear']=modeldata['Date'].dt.isocalendar().week.astype('int')
```

```python
modeldata[["Date", "Year", "Month", "Day", "WeekOfYear"]].head()
```

```python
modeldata.shape
```

* A year has four quarters:

  Q1 → Jan, Feb, Mar
  
  Q2 → Apr, May, Jun
  
  Q3 → Jul, Aug, Sep
  
  Q4 → Oct, Nov, Dec

```python
modeldata["Quarter"] = modeldata["Date"].dt.quarter
```

```python
modeldata[["Date", "Year", "Month", "Quarter", "WeekOfYear"]].head()
```

```python
modeldata.head()
```

#### Extracting meaningful information:

* CompetitionOpenSinceMonth, CompetitionOpenSinceYear are converted into months
* Promo2SinceWeek,Promo2SinceYear are converted into weeks or months

```python
modeldata["CompetitionOpenDate"] = pd.to_datetime(
    dict(
        year=modeldata["CompetitionOpenSinceYear"],
        month=modeldata["CompetitionOpenSinceMonth"],
        day=1
    ),
    errors="coerce"
)
```

```python
modeldata[["CompetitionOpenDate",'CompetitionOpenSinceYear','CompetitionOpenSinceMonth'] ]
```

```python
modeldata["CompetitionAge"] = (
    (modeldata["Date"] - modeldata["CompetitionOpenDate"]).dt.days / 30
).clip(lower=0)
```

```python
modeldata[["CompetitionAge","CompetitionOpenDate",'CompetitionOpenSinceYear','CompetitionOpenSinceMonth'] ].head()
```

### Create PromoDuration

```python
modeldata[
    [
        "Date",
        'Promo2',
        "Promo2SinceYear",
        "Promo2SinceWeek",
        
    ]
].head(10)
```

```python
modeldata.loc[
    modeldata["Promo2"] == 1,
    ["Promo2SinceYear", "Promo2SinceWeek"]
].head(10)
```

```python
from datetime import date

def get_promo_start(row):
    if row["Promo2"] == 0:
        return pd.NaT
    try:
        return pd.Timestamp(
            date.fromisocalendar(
                int(row["Promo2SinceYear"]),
                int(row["Promo2SinceWeek"]),
                1   # Monday
            )
        )
    except ValueError:
        return pd.NaT

modeldata["PromoStartDate"] = modeldata.apply(get_promo_start, axis=1)
```

```python
modeldata["PromoDuration"] = (
    (modeldata["Date"] - modeldata["PromoStartDate"]).dt.days / 30
).clip(lower=0)

modeldata["PromoDuration"] = modeldata["PromoDuration"].fillna(0)
```

```python
modeldata.loc[
    modeldata["Promo2"] == 1,
    ["Promo2SinceYear", "Promo2SinceWeek", "PromoStartDate", "PromoDuration"]
].head()
```

```python
modeldata.drop(
    columns=[
        "CompetitionOpenSinceMonth",
        "CompetitionOpenSinceYear",
        "CompetitionOpenDate",
        "Promo2SinceWeek",
        "Promo2SinceYear",
        "PromoStartDate",
        'Date'
    ],
    inplace=True
)
```
